```python
from obspy.core import *
from matplotlib import pyplot as plt

from scipy.stats import norm
import numpy as np
```

```python
data_path = '/Volumes/Hao/SOSAF_data/067/YA/'

sta_file = open('./event15/stidx_BF.txt', 'r')
sta_lines = sta_file.readlines()
sta_file.close()

sta_list = []
for sta_line in sta_lines:
    stas = sta_line.split()
    if stas[-1] == 'nan':
        stas.remove('nan')
    sta_list += stas
    
sta_all = open('./code/stations_YA.txt', 'r')
sta_lines = sta_all.readlines()
sta_all.close

sta_dict = {}
for sta in sta_lines:
    sta_id, stlo, stla = sta.split()[0:3]
    stlo, stla = float(stlo), float(stla)
    sta_dict[sta_id] = [stlo, stla]
```

```python
t, fs = 4, 500
t0 = UTCDateTime('2020-03-07T17:34:41.5Z')
data_z = []
for i in range(len(sta_list)):
    tr = read(data_path+'R'+sta_list[i]+'/*DPZ*')
    tr_z = tr.slice(t0-t,t0+t)
    tr_z = tr_z.filter('bandpass', freqmin=3, freqmax=5, zerophase=True)
    data = tr_z[0].data
    data_z.append(data)

amp = []
for i in range(len(data_z)):
    amp.append(np.sqrt(np.sum([k**2 for k in data_z[i]])))

R = 6371393 #m
dx, dy, dt, cc_all = [], [], [], []
for i in range(len(sta_list)):
    for j in range(i+1,len(sta_list)):
        st1, st2 = sta_list[i], sta_list[j]
        lo1, la1 = sta_dict[st1]
        lo2, la2 = sta_dict[st2]
        dx.append(R*((lo2-lo1)*np.pi/180)*np.cos((la2+la1)*np.pi/360))
        dy.append(R*(la2-la1)*np.pi/180)
        
        cc = np.correlate(data_z[i], data_z[j],'full')/ \
            (amp[i]*amp[j])
        cc_all.append(max(cc))
        dt.append((2*t*fs - np.argmax(cc))/fs) #t2 - t1


# plt.plot(cc)
```

```python
cc_lim = 0.8
dx_cc, dy_cc, dt_cc, cc_cc = [], [], [], []
for i in range(len(dx)):
    if cc_all[i]>cc_lim and abs(dt[i])<0.2 and dt[i]*dx[i]>0:
        dx_cc.append(dx[i])
        dy_cc.append(dy[i])
        dt_cc.append(dt[i])
        cc_cc.append(cc_all[i])

plt.figure(figsize=(12,10))
plt.scatter(dx_cc, dy_cc, c=dt_cc, marker='.', linewidths=4, cmap='jet')
plt.scatter([-i for i in dx_cc], [-i for i in dy_cc], c=[-i for i in dt_cc], marker='.', linewidths=4, cmap='jet')
plt.xticks(size=20)
plt.yticks(size=20)
plt.xlim([-100, 100])
plt.ylim([-100, 100])
plt.xlabel('W-E Distance (m)', fontsize=20)
plt.ylabel('N-S Distance (m)', fontsize=20)

plt.colorbar()

T = np.array(dt_cc)
T.resize(165,1)
D = np.array([dx_cc,dy_cc]).T
S = np.dot(np.linalg.pinv(D), T)

plt.arrow(0, 0, 18.8, 16.7, linewidth=4, head_width=1, head_length=1, fc='k', ec='k')
```

```python
t_s = 2
t_line = (np.arange(2*t_s*fs+1)-t_s*fs-1)/fs

plt.figure(figsize=(12,10))
for i in range(len(data_z[0:20])):
    plt.plot(t_line, data_z[i][(t-t_s)*fs:(t+t_s)*fs+1]/2000+i, 'k')
plt.xlabel('Time (s)', fontsize=20)
plt.ylabel('Station Index', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
    
z_stack = np.zeros(2*t_s*fs+1)
for i in range(len(data_z[0:102])):
    z_stack += data_z[i][(t-t_s)*fs:(t+t_s)*fs+1]

plt.figure(figsize=(12,10))
plt.plot(t_line, data_z[0][(t-t_s)*fs:(t+t_s)*fs+1]/2000, 'k')
for i in range(1, len(data_z[1:20])):
    ddt = dx[i-1]*S[0][0] +dy[i-1]*S[1][0]
    nt = int(ddt*fs)
    plt.plot(t_line, data_z[i][(t-t_s)*fs+nt:(t+t_s)*fs+nt+1]/2000+i, 'k')
plt.xlabel('Time (s)', fontsize=20)
plt.ylabel('Station Index', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks([0,5,10,15,20], fontsize=15)

z_stack_corr = np.zeros(2*t_s*fs+1)
for i in range(1, len(data_z[1:102])):
    ddt = dx[i-1]*S[0][0] +dy[i-1]*S[1][0]
    nt = int(ddt*fs)
    z_stack_corr += data_z[i][(t-t_s)*fs+nt:(t+t_s)*fs+nt+1]
```

```python
plt.figure(figsize=(12,8))
plt.plot(t_line, z_stack/102)
plt.plot(t_line, z_stack_corr/102)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Time (s)', fontsize=15)
plt.ylabel('Amplitude', fontsize=15)
plt.legend(['Raw', 'Corrected'], fontsize=15)
```

```python
data_n, data_e = [], []
for i in range(len(sta_list)):
    tr = read(data_path+'R'+sta_list[i]+'/*DPN*')
    tr_n = tr.slice(t0-t,t0+t)
    tr_n = tr_n.filter('bandpass', freqmin=3, freqmax=5, zerophase=True)
    data = tr_n[0].data
    data_n.append(data)
    
    tr = read(data_path+'R'+sta_list[i]+'/*DPE*')
    tr_e = tr.slice(t0-t,t0+t)
    tr_e = tr_e.filter('bandpass', freqmin=3, freqmax=5, zerophase=True)
    data = tr_e[0].data
    data_e.append(data)
```

```python
n_stack, e_stack = np.zeros(2*t_s*fs+1), np.zeros(2*t_s*fs+1)
for i in range(1, len(data_z[1:102])):
    ddt = dx[i-1]*S[0][0] +dy[i-1]*S[1][0]
    nt = int(ddt*fs)
    n_stack += data_n[i][(t-t_s)*fs+nt:(t+t_s)*fs+nt+1]
    e_stack += data_e[i][(t-t_s)*fs+nt:(t+t_s)*fs+nt+1]

kx, ky = S[0]/np.sqrt(S[0]**2+S[1]**2), S[1]/np.sqrt(S[0]**2+S[1]**2)
z_stack, n_stack, e_stack = z_stack_corr/102, n_stack/102, e_stack/102
r_stack = kx*e_stack + ky*n_stack
t_stack = kx*n_stack - ky*e_stack

plt.figure(figsize=(10,5))
plt.plot(t_line[750:1250], z_stack[750:1250], 'r')
plt.plot(t_line[750:1250], r_stack[750:1250], 'b')
plt.plot(t_line[750:1250], t_stack[750:1250], 'k')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Time (s)', fontsize=15)
plt.ylabel('Amplitude', fontsize=15)
plt.legend(['Z', 'R', 'T'], fontsize=15)


plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
plt.scatter(n_stack[750:1250], e_stack[750:1250], c=t_line[750:1250], marker='.')
plt.xlim([-6e2, 6e2])
plt.ylim([-6e2, 6e2])
plt.xticks([])
plt.yticks([])
plt.xlabel('E', fontsize=20)
plt.ylabel('N', fontsize=20)
plt.colorbar()

plt.subplot(1,3,2)
plt.scatter(r_stack[750:1250], z_stack[750:1250], c=t_line[750:1250], marker='.')
plt.xlim([-1e3, 1e3])
plt.ylim([-1e3, 1e3])
plt.xticks([])
plt.yticks([])
plt.xlabel('R', fontsize=20)
plt.ylabel('Z', fontsize=20)
plt.colorbar()

plt.subplot(1,3,3)
plt.scatter(t_stack[750:1250], z_stack[750:1250], c=t_line[750:1250], marker='.')
plt.xlim([-1e3, 1e3])
plt.ylim([-1e3, 1e3])
plt.xticks([])
plt.yticks([])
plt.xlabel('T', fontsize=20)
plt.ylabel('Z', fontsize=20)
plt.colorbar()
```
